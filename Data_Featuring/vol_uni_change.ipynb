```python
import sys
sys.path.append("/home/luke/Documents/GitHub/Data_Handling")

import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt

from Training_Features.realized_volatiliity import realized_vol
from Training_Features.unidirectional_change import maxLoss_list, maxProfit_list
```

```python
path = "/srv/sqc/volatility_exploration/NBBO_Candled/Most_Recent_Protocol/Stock_Data"
all_files = os.listdir(path)

for ind in range(len(all_files)):
    file = pd.read_csv(path + "/" + all_files[ind])
    ask_list = file["Ask"].tolist()

    lookback = 600
    padding = np.zeros(lookback)

    real_vol = realized_vol(ask_list, lookback)
    real_vol_input = np.concatenate((padding, real_vol))
    real_vol_output = np.concatenate((real_vol, padding))

    maxLoss = maxLoss_list(ask_list, lookback)
    maxProfit = maxProfit_list(ask_list, lookback)
    maxLoss_input = np.concatenate((padding, maxLoss))
    maxProfit_input = np.concatenate((padding, maxProfit))
    maxLoss_output = np.concatenate((maxLoss, padding))
    maxProfit_output = np.concatenate((maxProfit, padding))

    file["Real_Vol_input"] = real_vol_input
    file["Real_Vol_output"] = real_vol_output
    file["MaxLoss_input"] = maxLoss_input
    file["MaxProfit_input"] = maxProfit_input
    file["MaxLoss_output"] = maxLoss_output
    file["MaxProfit_output"] = maxProfit_output

    output_path = "/srv/sqc/volatility_exploration/NBBO_Candled/Most_Recent_Protocol/Real_Vol_Uni_Max/"
    file.to_csv(output_path + all_files[ind])
    print(ind)
```
